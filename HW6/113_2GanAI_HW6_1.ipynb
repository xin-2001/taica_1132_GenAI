## 5. 打造一個可以一直說下去的對話機器人

````python
system = "你是一個友善的對話機器人，負責回答使用者的提問，回答需根據受試者的提問層次，從簡單回答開始，逐步深化內容，避免一次性提供完整答案。回答方式需以提示或建議為主，讓受試者透過提問獲取關鍵資訊。回答時請使用台灣用語及繁體中文。"
description = "你好，我是你的 AI 聊天機器人, 什麼話都可以跟我聊哦 :)"
model = model
````

````python
icon = "(*ﾟ▽ﾟ*): " # 設置機器人回答的頭貼
messages = [{"role":"system", "content":system}]
print(icon + description + '\n')

while True:
    prompt = input('> ')
    if 'bye' in prompt:
        print('再見, 下次再聊!')
        break
    messages.append({"role": "user", "content": prompt})
    chat_completion = client.chat.completions.create(
        messages=messages,
        model=model,
        )

    reply = chat_completion.choices[0].message.content
    print(icon + reply)
    print()
    messages.append({"role": "assistant", "content": reply})
````

對話機器人 app 設定

````python
title = "柴柴引導式機器人"
description = "你好，我是你的 AI 聊天機器人, 什麼話都可以跟我聊哦 :)"
system = "你是一個友善的對話機器人，負責回答使用者的提問，回答需根據受試者的提問層次，從簡單回答開始，逐步深化內容，避免一次性提供完整答案。回答方式需以提示或建議為主，讓受試者透過提問獲取關鍵資訊。回答時請使用台灣用語及繁體中文。"
model = model
````

### 設置機器人的初始狀態

````python
initial_messages = [{"role":"system",
             "content":system},
            {"role":"assistant",
            'content':description}]
````

````python
state = gr.State(messages)
````

### 設置對話回覆的function

````python
def pipi(prompt, messages):
    messages.append({"role": "user", "content": prompt})
    chat_completion = client.chat.completions.create(
        messages=messages,
        model=model,
        )
    reply = chat_completion.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})
    #history = history + [[prompt, reply]]
    return messages, messages
````

````python
chatbot = gr.Chatbot(type="messages")
````

### 定義清除輸入框

````python
def clear_textbox(chatbot, state):
  return "", chatbot, state
````

- 在chatbot 中設置``` value=initial_messages``` 可以在建立聊天室後，由機器人先傳第一則訊息。
- 在```msg.submit```後設置```then(fn=clear_textbox, inputs=[chatbot, state], outputs=[msg, chatbot, state])``` ，可以在送出訊息後，自動將輸入框清空。
  > 其中```clear_textbox```函式是被定義用來清除輸入框的關鍵。



````python
with gr.Blocks(title=title) as demo:
  gr.Markdown(f"## 🤖 {title}\n{description}")
  #chatbot = gr.Chatbot(type="messages")
  chatbot = gr.Chatbot(type="messages", value=initial_messages) #進入聊天室後機器人先傳第一則訊息
  msg = gr.Textbox(label="輸入訊息")
  state = gr.State(initial_messages.copy())  # 務必用 copy()
  # 清除送出後輸入框內容
  msg.submit(fn=pipi, inputs=[msg, state], outputs=[chatbot, state]).then(fn=clear_textbox, inputs=[chatbot, state], outputs=[msg, chatbot, state])
  #msg.submit(fn=pipi, inputs=[msg, state], outputs=[chatbot, state])

demo.launch(share=True, debug=True)
````
