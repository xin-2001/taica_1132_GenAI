## 相關的數學函式

3. $\left(\left(x-\sin\left(a\pi\right)\cos\left(b\pi-0.5\pi\right)\right)^{2}+\left(y-\sin\left(b\pi-0.5\pi\right)\right)^{2}+\left(\frac{\operatorname{abs}\left(\cos\left(a\pi\right)\cos\left(b\pi-0.5\pi\right)\right)}{\cos\left(a\pi\right)\cos\left(b\pi-0.5\pi\right)}\sqrt{1-x^{2}-y^{2}}-\sqrt{\left(1-\sin\left(a\pi\right)^{2}\cos\left(b\pi-0.5\pi\right)^{2}-\sin\left(b\pi-0.5\pi\right)^{2}\right)}\right)^{2}-2\right)\sqrt{\left(\left(x-\sin\left(a\pi\right)\cos\left(b\pi\right)\right)^{2}+\left(y-\sin\left(b\pi\right)\right)^{2}+\left(\sqrt{1-x^{2}-y^{2}}-\sqrt{1-\sin\left(a\pi\right)^{2}\cos\left(b\pi\right)^{2}-\sin\left(b\pi\right)^{2}}\right)^{2}-0.1\right)}=0$
4. $\left(\left(x-\sin\left(a\pi\right)\cos\left(b\pi-0.5\pi\right)\right)^{2}+\left(y-\sin\left(b\pi-0.5\pi\right)\right)^{2}+\left(\frac{\operatorname{abs}\left(\cos\left(a\pi\right)\cos\left(b\pi-0.5\pi\right)\right)}{\cos\left(a\pi\right)\cos\left(b\pi-0.5\pi\right)}\sqrt{1-x^{2}-y^{2}}-\sqrt{\left(1-\sin\left(a\pi\right)^{2}\cos\left(b\pi-0.5\pi\right)^{2}-\sin\left(b\pi-0.5\pi\right)^{2}\right)}\right)^{2}-2\right)\sqrt{-\cos a\pi}\sqrt{-\left(\left(x-\sin\left(a\pi\right)\cos\left(b\pi\right)\right)^{2}+\left(y-\sin\left(b\pi\right)\right)^{2}+\left(\sqrt{1-x^{2}-y^{2}}-\sqrt{1-\sin\left(a\pi\right)^{2}\cos\left(b\pi\right)^{2}-\sin\left(b\pi\right)^{2}}\right)^{2}-0.1\right)}=0$

### 著色面積

## 主程式
### 說明
參考網路資源，找到使用函數製作的卡通圖案，將函式使用python呈現，並在置圖時加入參數增加其美觀，例如：線的粗細、填充的顏色等，以及匯入中文字體使其可以在圖表中呈現中文字。
這些函數組合成的結果會是一顆神奇寶貝球，除了是神奇寶貝球的圖案外，我還額外設置了a、b兩個參數，透過調整a、b參數，可以使神奇寶貝球呈現不同的視角。


````python
# 設定參數 a 和 b
"""
可自行調整
-0.5 < a < 0.5
-0.5 < b < 0.5
"""
a = -0.24
b = -0.20
````

````python
def equation1(x, y):
  return x**2 + y**2 - 1

def equation2(x, y, a, b):
  term1 = (x - np.sin(a * np.pi) * np.cos(b * np.pi))**2
  term2 = (y - np.sin(b * np.pi))**2
  term3 = (np.sqrt(1 - x**2 - y**2) - np.sqrt(1 - (np.sin(a * np.pi) * np.cos(b * np.pi))**2 - (np.sin(b * np.pi))**2))**2
  return np.sqrt(np.cos(a * np.pi)) * (term1 + term2 + term3 - 0.1)

def equation3(x, y, a, b):
  term1 = (x - np.sin(a * np.pi) * np.cos(b * np.pi - 0.5 * np.pi))**2
  term2 = (y - np.sin(b * np.pi - 0.5 * np.pi))**2
  term3_inner = (1 - np.sin(a * np.pi)**2 * np.cos(b * np.pi - 0.5 * np.pi)**2 - np.sin(b * np.pi - 0.5 * np.pi)**2)
  # 使用 np.where 避免除以 0 的錯誤
  term3 = (np.abs(np.cos(a * np.pi) * np.cos(b * np.pi - 0.5 * np.pi)) / (np.where(np.cos(a * np.pi) * np.cos(b * np.pi - 0.5 * np.pi) == 0, 1e-10, np.cos(a * np.pi) * np.cos(b * np.pi - 0.5 * np.pi))) * np.sqrt(1 - x**2 - y**2) - np.sqrt(np.where(term3_inner < 0, 0, term3_inner)))**2
  return (term1 + term2 + term3 - 2) * np.sqrt(equation2(x, y, a, b))


def equation4(x, y, a, b):
  term1 = (x - np.sin(a * np.pi) * np.cos(b * np.pi))**2
  term2 = (y - np.sin(b * np.pi))**2
  term3 = (np.sqrt(1 - x**2 - y**2) - np.sqrt(1 - (np.sin(a * np.pi) * np.cos(b * np.pi))**2 - (np.sin(b * np.pi))**2))**2
  return np.sqrt(np.cos(a * np.pi)) * (term1 + term2 + term3 - 0.05)

# 生成 x 和 y 的網格
x = np.linspace(-1.5, 1.5, 400)
y = np.linspace(-1.5, 1.5, 400)
X, Y = np.meshgrid(x, y)


# 繪製輪廓線
plt.figure(figsize=(8, 8))
plt.contour(X, Y, equation1(X, Y), levels=[0], colors='black', linewidths=2) # linewidths線的粗細
plt.contour(X, Y, equation2(X, Y, a, b), levels=[0], colors='black')
plt.contour(X, Y, equation3(X, Y, a, b), levels=[0], colors='black', linewidths=10)
plt.contour(X, Y, equation4(X, Y, a, b), levels=[0], colors='black')

plt.contourf(X, Y, equation1(X, Y), levels=[-np.inf, 0], colors='white', alpha=0.5)  # levels=[-np.inf, 0] 表示填充 equation1(X, Y) <= 0 的區域
plt.contourf(X, Y, equation2(X, Y, a, b), levels=[-np.inf, 0], colors='black', alpha=0.8)

Z3 = equation3(X, Y, a, b)
plt.contourf(X, Y, Z3, levels=[0, np.inf], colors='red', alpha=0.8)  # 填充 Z3 > 0 的區域(虹色部分)

plt.contourf(X, Y, equation4(X, Y, a, b), levels=[-np.inf, 0], colors='white', alpha=1)

plt.xlabel("x")
plt.ylabel("y")
plt.title("神奇寶貝球")
# plt.grid(True) #網格
#plt.legend()
plt.show()
````
